                                                        RENAMING COLUMNS

```python
data.rename(columns={'Year (1)': 'Year', 'Area Affected m.ha (2)': 'Area_Affected', 'Population Affected in Million (3)':'Population_Affected','Damage to Crops - Area m.ha (4)':'DamagedCrops_Area','Damage to Crops - Value in Rs. Crore (5)':'DamagedCrops_Value','Damage to Houses - Numbers (6)':'DamagedHouses_Numbers','Damage to Houses - Value in Rs. Crore (7)':'DamagedHouses_Value','Cattle Lost Numbers (8)':'Cattle_Lost','Human Lives Lost Numbers (9)':'HumanLives_Lost','Damage to Public Utilities in Rs. Crore (10)':'DamagedPublic_Utilities','Total Damages Crops, Houses & Public Utilities in Rs. Crore (11) = (Col.5+7+10)':'Total_Damage',}, inplace=True)
```

                                                    FINDING MISSING VALUES

```python
total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum()/data.isnull().count()).sort_values(ascending=False)
missing = pd.concat([total,percent], axis=1, keys=['Total','Percent'])
missing.head(20)
```

```python
data.isnull().sum()
```

                                                    HANDLING MISSING VALUES

```python
area_mode = data.Area_Affected.mode()[0]
data.Area_Affected.fillna(area_mode, inplace=True)

pop_mode = data.Population_Affected.mode()[0]
data.Population_Affected.fillna(pop_mode, inplace=True)

carea_mode = data.DamagedCrops_Area.mode()[0]
data.DamagedCrops_Area.fillna(carea_mode, inplace=True)

cvalue_mode = data.DamagedCrops_Value.mode()[0]
data.DamagedCrops_Value.fillna(cvalue_mode, inplace=True)

dhouse_mode = data.DamagedHouses_Numbers.mode()[0]
data.DamagedHouses_Numbers.fillna(dhouse_mode, inplace=True)

dvalue_mode = data.DamagedHouses_Value.mode()[0]
data.DamagedHouses_Value.fillna(dvalue_mode, inplace=True)

cattle_mode = data.Cattle_Lost.mode()[0]
data.Cattle_Lost.fillna(cattle_mode, inplace=True)

live_mode = data.HumanLives_Lost.mode()[0]
data.HumanLives_Lost.fillna(live_mode, inplace=True)

pu_mode = data.DamagedPublic_Utilities.mode()[0]
data.DamagedPublic_Utilities.fillna(pu_mode, inplace=True)

total_mode = data.Total_Damage.mode()[0]
data.Total_Damage.fillna(total_mode, inplace=True)

data.isnull().sum()
```
